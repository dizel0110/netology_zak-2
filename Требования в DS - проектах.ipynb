Продукт компании “Нетология” - профессия Data Scientist является одним из продуктов линейки образовательных онлайн-курсов. Данная профессия обучает специалистов в нескольких направлениях, в результате обучения мы имеем Junior-специалистов широких специальностей. И, особенность данного продукта-профессии, - все выпускники так или иначе являются потенциальными участниками data-driven подхода. В этом и соль данного задания, мы находимся в самом центре обсуждаемой темы.

#### Аналитика по данному продукту:

Важно не только привлечь потенциального слушателя профессии, но и следить за успехами и неудачами конкретных пользователей. Смотреть уровень осведомленности и качество профессионального роста.  Проанализировать количество людей, способных на выполнение задач разного уровня сложности. Ввести критерий сложности задач и времени на их выполнения, оригинальности подхода. Конечно, есть нюансы и люди-инсайты. Но это отдельные персонажи, живущие по-особым законам. Точно не бегемоты, но и не большинство беспристрастных слушателей профессии. 

#### Ключевые метрики:

Профессия, как элемент онлайн образовательного проекта.

Я бы выделил две ключевых метрики: 

Метрика привлечения пользователей:
MAU, monthly active users - количество активных пользователей в месяц.

Метрика использования продукта:
ER, engagement rate — коэффициент вовлечённости.

Данные метрики говорят о потенциале слушателей, о способности пройти все круги профессии Data Scientist.


Можно добавить ещё некую бизнес- пользу, ввести некое подобие сальдо.

Метрика монетизации:

ARPPU — средняя прибыль от одного платящего пользователя за определённый срок.

CAC, customer acquisition cost — стоимость привлечения клиента — сколько денег в среднем нужно потратить на рекламу и продвижение, чтобы привлечь одного пользователя.

Случайные люди в данную профессию не пойдут. Это не точно, но факт. Сначала надо сформулировать, что в данном продукте является инсайтом. Я немного был ошарашен новой трансформацией самой Нетологии, стоило только отвлечься на построения стойл (реальная причина отсутствия на первых двух лекциях), как ромашка превратилась в четыре кружочка, страшная картина. Человек привыкает к одному, а через некоторое время ему показывают другое. Но Мы то знаем, что всего цветов три, и они  другого оттенка. Хорошо, что ещё факультеты не подкрасили. Инсайт можно словить в любом месте, вовлечённость должна присутствовать не только у слушателей курса, но и у профессорского состава, координаторов и помощников чата. Стоит проанализировать в дополнении к вышесказанному количество сданных работ с первого раза, количество возвращённых работ. Цифры беспристрастны к болям студентов, преподавателей и  заказчиков интеллектуальных трудовых единиц для работы на фронте data-driven. И хотелось бы вывести критерий оптимальной траектории обучения, с возможностью накопления нестандартных решений всевозможных учебно-практических задач. Инсайтом в данном случае может служить озарение (просветление) при разборе домашних заданий, и, желательно, взаимно для участников.
